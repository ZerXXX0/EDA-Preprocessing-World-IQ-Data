```python
#import
import statistics
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/ZerXXX0/EDA-Preprocessing-World-IQ-Data/main/avgIQpercountry.csv')
df = pd.DataFrame(data)
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
sns.pairplot(df, hue='Average IQ')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Average IQ'],bins=20,kde=True)
plt.title("Histogram")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Continent', y='Average IQ')
plt.title("Boxplot")
plt.show()
```

```python
correlation_matrix = df[['Average IQ','HDI (2021)']].corr()
print(correlation_matrix)
```

```python
correlation_matrix1 = df[['Average IQ',' GNI - 2021']].corr()
print(correlation_matrix)
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Heatmap Correlation")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(correlation_matrix1, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Heatmap Correlation")
plt.show()
```
